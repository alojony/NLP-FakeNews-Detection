```python
#Imports

from textblob import TextBlob
import pandas as pd
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
import numpy as np
nltk.download('vader_lexicon')

# Initialize VADER
sid = SentimentIntensityAnalyzer()
```

```python
#Load data

#Path to Excel file
file_path = '/Users/oliviamirijello/Desktop/TextMining/Project/NLP-FakeNews-Detection/archive/truth_seeker.xlsx'

#Read Excel file
df = pd.read_excel(file_path)
```

```python
# Using TextBlob to obtain polarity scores 

def analyze_sentiment(text):
    polarity = TextBlob(str(text)).sentiment.polarity
    if polarity > 0:
        sentiment = 'positive'
    elif polarity < 0:
        sentiment = 'negative'
    else: 
        sentiment = 'neutral'
    return polarity, sentiment

#Apply the function to the tweet colymn and create 2 new columns for the results 
df[['polarity_score', 'tone']] = df['tweet'].apply(lambda x: pd.Series(analyze_sentiment(x)))

#Display the first few rows to verify the new columns
print(df[['tweet', 'polarity_score', 'tone']].head())
```

```python
#Showing the unique values of tone to see that we have positive, negative, and neutral entries 
print(df['tone'].unique())
```

```python
#Frequency table to explore split between positive, negative, and neutral 
tone_freq = df['tone'].value_counts()
print(tone_freq)
```

```python
#Load data

#Path to Excel file
file_path = '/Users/oliviamirijello/Desktop/TextMining/Project/NLP-FakeNews-Detection/archive/truth_seeker.xlsx'

#Read Excel file
dfV = pd.read_excel(file_path)

#Remove NA from tweet
dfV = df.dropna(subset=['tweet'])
```

```python
# Using VADER to obtain polarity scores 

# Apply SID to each tweet and expand the results into separate columns
dfV[['neg', 'neu', 'pos', 'compound']] = dfV['tweet'].apply(lambda x: pd.Series(sid.polarity_scores(x)))

#Display the first few rows to verify the new columns
print(dfV[['tweet', 'neg', 'neu', 'pos', 'compound']].head())
```

```python
# Merge the dataframes on the 'tweet' column
merged_df = pd.merge(df, dfV, on='tweet', how='inner')
```

```python
# Calculate the difference between 'polarity_score' and 'compound'
merged_df['score_difference'] = merged_df['polarity_score_y'] - merged_df['compound']

# Adding column for absolute value of compound entries 
merged_df['absolute_compound'] = merged_df['compound'].abs()

# Select the relevant columns to display
result_df = merged_df[['polarity_score_y', 'compound', 'absolute_compound','score_difference']].copy()

# Print the results
print(result_df)
```

```python
# Here we can see that VADER gives more accurate results (not just 0.00). 
```

```python
#User Demographic Analysis: analyze sentiment by different demographic groups
#Analyze which authors tweet the most
tweet_counts = dfV['author'].value_counts()

# Plotting tweet counts
plt.figure(figsize=(10, 6))
tweet_counts.plot(kind='bar')
plt.title('Number of Tweets by Author')
plt.xlabel('Author')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=45)
plt.show()
```

```python
#Group by the authors and then calculate the mean polarity score
sentiment_distribution = dfV.groupby('author')['compound'].mean()

# Plotting sentiment distribution
plt.figure(figsize=(10, 6))
sentiment_distribution.plot(kind='bar', color='skyblue')
plt.title('Average Sentiment Polarity by Author')
plt.xlabel('Author')
plt.ylabel('Average Polarity Score')
plt.axhline(0, color='gray', linewidth=0.5)
plt.xticks(rotation=45)
plt.show()
```

```python
# Finding the author with the maximum polarity score
max_polarity_author = dfV.loc[dfV['compound'].idxmax(), 'author']

# Finding the author with the minimum polarity score
min_polarity_author = dfV.loc[dfV['compound'].idxmin(), 'author']

print("The author with the highest polarity score is ",max_polarity_author," and the author with the lowest polarity score is ",min_polarity_author,".")
```

```python
#Showing the distribution of the polarity in tweets by max and min 

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Filter the dataframe for the selected author
author_df1 = dfV[dfV['author'] == "Ciara O'Rourke"]
author_df2 = dfV[dfV['author'] == "Jon Greenberg"]

# Plot the distribution of polarity scores for the selected author
ax1.hist(author_df1['compound'], bins=20, color='pink', edgecolor='black')
ax1.set_title("Distribution of Polarity Scores for Ciara O'Rourke")
ax1.set_xlabel('Polarity')
ax1.set_ylabel('Frequency')
ax1.grid(axis='y',alpha=0.75)

ax2.hist(author_df2['compound'], bins=20, color='skyblue', edgecolor='black')
ax2.set_title("Distribution of Polarity Scores for Jon Greenberg")
ax2.set_xlabel('Polarity')
ax2.set_ylabel('Frequency')
ax2.grid(axis='y',alpha=0.75)

plt.tight_layout()
plt.show()
```

```python
# To find the author with the highest percentage of tweets

# Count the number of tweets per author
tweets_per_author = dfV['author'].value_counts(normalize=True) * 100  # This will give the percentage directly

# Identify the author with the highest percentage of tweets
author_with_most_tweets = tweets_per_author.idxmax()
max_percentage = tweets_per_author.max()

(author_with_most_tweets, max_percentage)
```

```python
#Finding skewness within polarity scores for each author

# Calculate skewness for each author
polarity_skewness = dfV.groupby('author')['polarity score'].skew()

# Find the author with the most left-skewed distribution
most_left_skewed_author = polarity_skewness.idxmin()

# Find the author with the most right-skewed distribution
most_right_skewed_author = polarity_skewness.idxmax()

# Output the results
polarity_skewness, most_left_skewed_author, most_right_skewed_author
```
